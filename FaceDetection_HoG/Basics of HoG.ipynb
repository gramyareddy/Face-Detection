# Basics of Histogram of Oriented Gradients

Gradients refer to the change in pixel values with respect to x,y direction.

Orientation refer to the direction of change .

Histogram denotes the frequency of the gradients in some direction

```python
from skimage.io import imread, imshow
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt
import cv2
import numpy as np
print("imported")
```

```python
# Read image

img = imread('C:\\MyLearnings\\DeepLearning\\Facial Projects\\Face Detection\\HoGImage.jpg')
imshow(img)


```

# Feature Descriptors
The hog function that calculates gradients,magnitude and orientation

1.The orientations are the number of buckets we want to create.

2.pixels_per_cell defines the size of the cell for which we create the histograms.

3.We have another hyperparameter cells_per_block which is the size of the block over which we normalize the histogram. Here, we mention the cells per blocks and not the number of pixels. So, instead of writing 16 x 16, we will use 2 x 2 here

```python
#creating hog features 
fd, hog_image = hog(img, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=True, multichannel=True)
print("HoG created")
```
